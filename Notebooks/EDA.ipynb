```python
import pandas as pd

# Load the dataset
df = pd.read_csv('C:\\Users\\Admin\\Downloads\\diabetes.csv')  # Use double slashes \\ or raw string


# Show basic info
print(df.shape)     # Rows and columns
print(df.columns)   # List of column names
df.head()           # Show first 5 rows

```

```python
# Check dataset shape (rows, columns)
print("Shape:", df.shape)

# View column names
print("Columns:", df.columns.tolist())

# View data types and missing info
df.info()

# View summary statistics
df.describe()
```

```python
# List of columns to fix
cols_to_fix = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

# Replace 0s with median of each column
for col in cols_to_fix:
    median = df[col].median()
    df[col] = df[col].replace(0, median)
df.to_csv("diabetes_cleaned.csv", index=False)

print("✅ Cleaned dataset saved as 'diabetes_cleaned.csv'")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# List of cleaned features
cleaned_columns = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

# Plot each distribution in a loop
for col in cleaned_columns:
    plt.figure(figsize=(7, 4))
    sns.histplot(df[col], kde=True, bins=30, color='skyblue', edgecolor='black')
    plt.title(f'{col} Distribution After Cleaning')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
sns.countplot(x='Outcome', data=df, palette='Set2')
plt.title("Target Class Distribution")
plt.xlabel("Diabetes Outcome")
plt.ylabel("Count")
plt.show()

print(df['Outcome'].value_counts(normalize=True))  # Show class proportion
```

```python
 plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
for col in ['Glucose', 'BloodPressure', 'BMI', 'Insulin', 'Age']:
    plt.figure(figsize=(7, 4))
    sns.boxplot(x='Outcome', y=col, data=df, palette='Set1')
    plt.title(f'{col} by Diabetes Outcome')
    plt.show()
```

```python
sns.pairplot(df[['Glucose', 'BMI', 'Age', 'DiabetesPedigreeFunction', 'Outcome']], hue='Outcome', palette='husl')
plt.suptitle("Pairwise Feature Relationships", y=1.02)
plt.show()
```
